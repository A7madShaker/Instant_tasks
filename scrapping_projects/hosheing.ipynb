```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import csv

# إعداد متصفح Selenium
options = webdriver.ChromeOptions()
# options.add_argument('--headless')  # أزل التعليق إذا تريد تشغيل بدون واجهة
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=options)

# إعداد ملف CSV
csvfile = open("house.csv", "w", encoding="utf-8-sig", newline='')
writer = csv.writer(csvfile)
writer.writerow(["الموقع", "اسم الشقة", "السعر"])  # رأس الجدول

try:
    # الانتقال إلى صفحة تأجير الشقق في Properstar
    url = "https://www.properstar.qa/الولايات-المتحدة/عقارات-للايجار/شقة"
    driver.get(url)
    time.sleep(5)

    # التمرير لتحميل المزيد من الشقق
    for _ in range(2):
        driver.execute_script("window.scrollTo(0, document.documentElement.scrollHeight);")
        time.sleep(2)

    # الحصول على مصدر الصفحة
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")

    # البحث عن قوائم الشقق
    listings = soup.find_all("div", class_="item-data")

    if not listings:
        print("لم يتم العثور على شقق. تحقق من هيكلية الصفحة أو الاتصال بالإنترنت.")
        print("مصدر الصفحة:", page_source[:1000])
    else:
        print("الشقق الموجودة:")
        for listing in listings:
            # استخراج اسم الشقة
            title_elem = listing.find("a", class_="link listing-title stretched-link")
            title = title_elem.text.strip() if title_elem else "غير متوفر"

            # استخراج السعر
            price_elem = listing.find("div", class_="listing-price")
            price = price_elem.text.strip() if price_elem else "غير متوفر"

            # استخراج الموقع
            location_elem = listing.find("div", class_="item-location")
            location = location_elem.text.strip() if location_elem else "غير متوفر"

            # طباعة البيانات
            print(f"اسم الشقة: {title}")
            print(f"السعر: {price}")
            print(f"الموقع: {location}")
            print("-" * 50)

            # كتابة البيانات في ملف CSV
            writer.writerow([location, title, price])

finally:
    # إغلاق ملف CSV والمتصفح
    csvfile.close()
    driver.quit()
```
